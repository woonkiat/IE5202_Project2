# Exponential Smoothing

## Exponential Weighted Moving Average

```python
fig, ax = plt.subplots(1,3,figsize=(15,5))
al = 0.1;
ewma = df_train['TrafficVolume'].ewm(alpha=al, min_periods=0)
ax[0].scatter(x=range(0,df_train.shape[0]), y=df_train['TrafficVolume'])
ax[0].scatter(x=range(0,df_train.shape[0]), y=ewma.mean(), color='red')

al = 0.3;
ewma = df_train['TrafficVolume'].ewm(alpha=al, min_periods=0)
ax[1].scatter(x=range(0,df_train.shape[0]), y=df_train['TrafficVolume'])
ax[1].scatter(x=range(0,df_train.shape[0]), y=ewma.mean(), color='red')

al = 0.5;
ewma = df_train['TrafficVolume'].ewm(alpha=al, min_periods=0)
ax[2].scatter(x=range(0,df_train.shape[0]), y=df_train['TrafficVolume'])
ax[2].scatter(x=range(0,df_train.shape[0]), y=ewma.mean(), color='red')
```

```python
alpha = np.linspace(0.01,1,num=100)
err = [];
for al in alpha:
    ewma = df_train['TrafficVolume'].ewm(alpha=al, min_periods=0)
    pred = ewma.mean();
    diff = df_train['TrafficVolume'] - pred.shift(1);
    err.append(np.sqrt((diff ** 2).mean()))
    
plt.plot(alpha, err)
optal = alpha[np.argmin(err)]
plt.axvline(x=optal, color='red')
print(optal,min(err))
```

```python
# given a series and alpha, return series of smoothed points
def double_exponential_smoothing(series, alpha, beta, L0, B0):
    result = []
    # No prediction for the first value
    result.append(np.nan)
    for n in range(0, len(series)):
        val = series[n]
        if n==0:
            level = alpha*val + (1-alpha)*(L0+B0);
            trend = beta*(level-L0) + (1-beta)*B0;
            last_level = level;
        else:
            level = alpha*val + (1-alpha)*(last_level+trend)
            trend = beta*(level-last_level) + (1-beta)*trend
            last_level = level;
            
        result.append(level+trend)
        
    # No prediction for the (n+1)th value
    return result[:-1]
```

```python
# Find the optimal alpha and beta
alpha = np.linspace(0.01,1,num=10)
beta = np.linspace(0.01,1,num=10)

series = df_train['TrafficVolume']
err = [];
min_err = 1000000;
for al in alpha:
    for be in beta:
        holt = double_exponential_smoothing(series, al, be, series[0], series[1]-series[0])
        holt = pd.DataFrame({"holt": holt}, index=df_train.index)
        pred = holt.dropna().holt.values;
        diff = series[1:] - pred;
        error = np.sqrt((diff ** 2).mean())
        if min_err > error:
            min_err = error
            alpha_hat = al
            beta_hat = be

        err.append(error)
        
print(alpha_hat, beta_hat)
```

## SARIMA model on residual

```python
resid = df_train['TrafficVolume'] - y_pred

fig, ax = plt.subplots(1,2,figsize=(12,6))
ax[0].scatter(df_train['time'][:1000], resid[:1000])
ax[1].scatter(df_train['time'][:100], resid[:100])
```

### clearly non-stationary, first take a seasonal difference of 24hr

```python
data = resid - resid.shift(24)
fig, ax = plt.subplots(1,2,figsize=(12,6))
ax[0].scatter(df_train['time'][:1000], data[:1000])
ax[1].scatter(df_train['time'][:100], data[:100])
```

### this data also appear to be non-stationary, take additional first difference

```python
data = resid - resid.shift(1) - (resid.shift(24) - resid.shift(25))
fig, ax = plt.subplots(1,2,figsize=(12,6))
ax[0].scatter(df_train['time'][:1000], data[:1000])
ax[1].scatter(df_train['time'][:100], data[:100])
```

```python
import statsmodels.graphics.tsaplots as sgt
lags = 100
alpha= 0.05
data = data.dropna()
fig, ax = plt.subplots(1,2,figsize=(12,6))
fig = sgt.plot_acf(data, ax=ax[0], lags=lags, alpha=alpha, unbiased=True)
fig = sgt.plot_pacf(data, ax=ax[1], lags=lags, alpha=alpha, method='ols')
```

### exponential decay in seasonal lags of the PACF <br> significant spike at lag 24,48 in the ACF suggesting seasonal MA(2) <br> non-seasonal spikes dies down for both ACF and PACF suggesting an ARMA model, start off with ARMA(1,1) <br> hence, begin with ARIMA(1,1,1)(0,1,2)24


```python
data = resid[-300*24:]
```

```python
import statsmodels.api as sm
model = sm.tsa.SARIMAX(data, trend='n', order=(1,1,1), seasonal_order=(0,1,2,24))
result_sm = model.fit()
print(result_sm.summary())
```

```python
pred = result_sm.predict(dynamic=False)
fig = plt.figure(figsize=(12,5))
plt.plot(data, label='Original')
plt.plot(pred, color='red', label='fitted')
plt.legend(loc='best')
```

```python
#zoom in to 1 month

fig = plt.figure(figsize=(12,5))
plt.plot(data[-24*28:], label='Original')
plt.plot(pred[-24*28:], color='red', label='fitted')
plt.legend(loc='best')
```
